```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
sns.displot(df['age'], kde=True)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.displot(df['age'], bins=35, color='red');
```

```python
sns.countplot(x='sex', data=df);
```

```python
sns.countplot(x='cp', data=df, hue='sex', palette='terrain');
plt.xlabel('hai')
```

```python
df.head()
```

```python
sns.boxplot(x = 'sex', y = 'age', data = df)
# x : categorical, y: continuous
```

```python
sns.boxplot(x='target', y='thalach', data=df, hue='sex');
```

```python
sns.scatterplot(x='chol', y='trestbps', data=df, 
                hue='sex', palette='Dark2', size='age')
```
